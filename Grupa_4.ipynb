```python
fig, axes = plt.subplots(3, 1, figsize=(6, 15))

fig.suptitle('Częstotliwość występowania pierwiastków', size=12, y=0.90)

swarm_iron = sns.boxplot(ax=axes[0], x='Iron', y='Type', data=glass)
plt.setp(swarm_iron.get_xticklabels(), rotation=90)
swarm_iron.set_xlabel('Iron')

swarm_barium = sns.boxplot(ax=axes[1], x='Barium', y='Type', data=glass)
plt.setp(swarm_barium.get_xticklabels(), rotation=90)
swarm_barium.set_xlabel('Barium')

swarm_calcium = sns.boxplot(ax=axes[2], x='Calcium', y='Type', data=glass)
plt.setp(swarm_calcium.get_xticklabels(), rotation=90)
swarm_calcium.set_xlabel('Kopytko')
```

```python
plt.style.use('ggplot')
```

```python
pogrupowano = glass.groupby('Type')['Sodium',	'Magnesium',	'Aluminum',	'Silicon',	'Potassium',	'Calcium',	'Barium',	'Iron'].aggregate('mean')
print(pogrupowano)

fig, axes = plt.subplots(2, 3, figsize=(30, 20))

szk_bud_p = pogrupowano.iloc[0].plot.pie(ax = axes[0, 0], autopct='%.0f%%')
szk_bud_p.title.set_text('Szkło budowlane przetworzone')
szk_bud_p.set_ylabel('')
szk_bud_p.title.set_size(20)
szk_bud_np = pogrupowano.iloc[1].plot.pie(ax = axes[0, 1], autopct='%.0f%%')
szk_bud_np.title.set_text('Szkło budowlane nieprzetworzone')
szk_bud_np.set_ylabel('')
szk_bud_np.title.set_size(20)
szk_poj = pogrupowano.iloc[2].plot.pie(ax = axes[0, 2], autopct='%.0f%%')
szk_poj.title.set_text('Szkło pojazdowe przetworzone')
szk_poj.set_ylabel('')
szk_poj.title.set_size(20)
cont = pogrupowano.iloc[3].plot.pie(ax = axes[1, 0], autopct='%.0f%%')
cont.title.set_text('Pojemniki')
cont.set_ylabel('')
cont.title.set_size(20)
zastawa = pogrupowano.iloc[4].plot.pie(ax = axes[1, 1], autopct='%.0f%%')
zastawa.title.set_text('Zastawa')
zastawa.set_ylabel('')
zastawa.title.set_size(20)
refl = pogrupowano.iloc[5].plot.pie(ax = axes[1, 2], autopct='%.0f%%')
refl.title.set_text('Reflektory')
refl.set_ylabel('')
refl.title.set_size(20)
```
